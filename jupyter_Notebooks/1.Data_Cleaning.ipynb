```python
df['Order_Date']=pd.to_datetime(df['Order_Date'],format='%d-%m-%Y')
df['Ship_Date']=pd.to_datetime(df['Ship_Date'],format='%d-%m-%Y')
```

```python
df[['Order_Date','Ship_Date']].dtypes
```

```python
bad_dates = df[df['Ship_Date']<df['Order_Date']]
bad_dates.shape
```

```python
bad_dates.head()
```

```python
text_cols=[
    'Gender','City','State','Country','Product_Name','Category',
    'Payment_Method','Ship_Mode','Order_Status'
]
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
df['Profit_Margin_Check'] = (df['Profit']/df['Sales'])*100
mismatch = df[(df['Profit_Margin_Check']-df['Profit_Margin']).abs()>0.1]
mismatch.shape
```

```python
mismatch.head()
```

```python
df['Profit_Margin']=df['Profit_Margin_Check']
```

```python
df.drop(columns=['Profit_Margin_Check'], inplace=True)
```

```python
df['Profit_Margin_Check'] = (df['Profit'] / df['Sales']) * 100
mismatch = df[(df['Profit_Margin_Check'] - df['Profit_Margin']).abs() > 0.1]
mismatch.shape
```

```python
df[['Sales','Profit']].describe()
```

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3-Q1

outliers_sales = df[(df['Sales']<Q1 - 1.5*IQR) | (df['Sales']>Q3 + 1.5 * IQR)]
outliers_sales.shape
```

```python
Q1p = df['Profit'].quantile(0.25)
Q3p = df['Profit'].quantile(0.75)
IQRp = Q3p - Q1p

outliers_profit = df[(df['Profit'] < Q1p - 1.5 * IQRp) | (df['Profit'] > Q3p + 1.5 * IQRp)]
outliers_profit.shape
```

```python
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Month'] = df['Order_Date'].dt.year

df['Order_Month_Name'] = df['Order_Date'].dt.strftime('%B')

df['Days_To_Ship'] = (df['Ship_Date']-df['Order_Date']).dt.days
```

```python
df.to_csv("Ecommerce_Cleaned.csv",index=False)
print("Cleaned File Saved as Ecommerce_Cleaned.csv")
```
