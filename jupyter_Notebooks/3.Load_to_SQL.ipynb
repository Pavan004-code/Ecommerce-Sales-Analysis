```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("Ecommerce_cleaned.csv",parse_dates =['Order_Date','Ship_Date'])
df.head()
```

```python
from sqlalchemy import create_engine

USERNAME = "root"
PASSWORD = "2004"
HOST = "localhost"
PORT ="3306"

engine_root = create_engine(f"mysql+pymysql://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/")
```

```python
from sqlalchemy import text

with engine_root.connect() as conn:
    conn.execute(text("CREATE DATABASE IF NOT EXISTS ecommerce;"))
    print("Database 'ecommerce' is ready.")
    
```

```python
engine = create_engine(f"mysql+pymysql://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/ecommerce")
```

```python
df.to_sql('orders_cleaned',con=engine, if_exists='replace',index=False,chunksize = 1000)
print("Data uploaded to Mysql table 'orders_cleaned'.")
```

```python
query = """
SELECT 
    COUNT(*) AS total_orders,
    SUM(Sales) AS total_sales,
    SUM(Profit) AS total_profit
FROM orders_cleaned;
"""

test_result = pd.read_sql(query, con=engine)
test_result
```
